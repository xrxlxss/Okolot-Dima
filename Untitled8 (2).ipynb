```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
x=np.array([-4.,-3.,0.,1.], dtype=float)
y=np.array([-7.,10.,-11.,-22.], dtype=float)
x2 =np.array([-2.,-0.5,0.5,2], dtype=float)
def lagrange_interpolation(x, y, x_test):
    n = len(x)
    p = np.zeros(n)
    for i in range(n):
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_test - x[j])/(x[i] - x[j])
        p[i] = p_i
    return np.dot(y, p)
for n in x2:
    fin = lagrange_interpolation(x, y, n)
    print("Значення функції у точці ",n," = ", fin.round(4))
```

```python
xnew=np.linspace(np.min(x),np.max(x),100)
ynew=[lagrange_interpolation(x,y,i) for i in xnew]

plt.plot(x, y, 'o', color='pink')
plt.plot(xnew, ynew, color='y')

plt.title(' Поліграм Лагранжа_1')

plt.grid(True)
plt.xlabel('x')

plt.ylabel('y')

plt.show()
```

```python
f = lagrange(x, y)

fig = plt.figure(figsize = (7,5))

plt.plot(xnew, f(xnew), 'pink', x, y, 'bo')

plt.title('Поліграм Лагранжа_2')

plt.grid()

plt.xlabel('x')

plt.ylabel('y')

plt.show()
```
